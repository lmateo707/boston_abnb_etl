```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "reviews.csv"

# Read Purchasing File and store into Pandas data frame
review_data = pd.read_csv(file_to_load,encoding="utf8")
```

```python
review_data.head()
```

```python
review_data.count()
```

```python
review_clean = review_data.dropna()
review_clean.count()
```

```python
#drop comments
review_cln = review_clean.drop(['comments'],axis=1)
#convert to datetime
# convert to datetime
review_cln["date"]=pd.to_datetime(review_cln["date"])
review_cln
```

```python
#groupby listing_id
review_grouped = review_cln.groupby(['listing_id'])
review_grouped.head()
```

```python
#for fun, lets do some counts
review_count = review_grouped.count()
review_count.sort_values(by=['id'],ascending=False)
```
